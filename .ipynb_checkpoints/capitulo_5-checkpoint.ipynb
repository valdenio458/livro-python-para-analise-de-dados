### Descartando entradas de um eixo

Com Series

```python
obj = pd.Series(np.arange(5.), index=['a', 'b', 'c', 'd', 'e'])
```

```python
obj
```

```python
new_obj = obj.drop('c')
```

```python
new_obj
```

```python
obj.drop(['d', 'c'])
```

Com DataFrame

```python
data = pd.DataFrame(np.arange(16).reshape((4,4)),
                    index=['Ohio', 'Colorado', 'Utah', 'New York'],
                    columns=['one', 'two', 'three', 'four'])
```

```python
data
```

```python
data.drop(['Colorado','Ohio'])
```

```python
data.drop('two', axis=1)
```

```python
data.drop(['two', 'four'], axis='columns')
```

```python
# Funções como drop são capazes de alterar a Series ou Dataframe in place, para isto
# acrescente inplace=True no segundo argumento de drop.
obj.drop('c', inplace=True)
```

```python
obj
```

### Seleção com loc e iloc

```python
data = pd.DataFrame(np.arange(16).reshape((4,4)),
                    index=['Ohio', 'Colorado', 'Utah', 'New York'],
                    columns=['one', 'two', 'three', 'four'])
```

```python
data
```

```python
# Selecionar uma única linha e várias colunas pelo rótulo
data.loc['Colorado', ['two', 'three']]
```

```python
# Seleções com inteiros usando iloc
data.iloc[2, [3, 0, 1]] # linha 2, colunas 3, 0, 1
```

```python
data.iloc[2] # igual a data.loc['Utah']
```

```python
data.iloc[[1, 2], [3, 0, 1]]
```

### Ordenação e classificação

```python
obj = pd.Series(range(4), index=['d','a', 'b', 'c'])
```

```python
obj
```

```python
obj.sort_index()
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2, 4)),
index=['three', 'one'],
columns=['d', 'a', 'b','c'])
```

```python
frame
```

```python
frame.sort_index()
```

```python
frame.sort_index(axis=1)
```

```python
# Ordenando em ordem decrescente
frame.sort_index(axis=1, ascending=False)
```

```python
# para ordenar uma Series de acordo com seus valores, utilize o método sort_values
obj = pd.Series([4, 7, -3, 2])
```

```python
obj
```

```python
obj.sort_values()
```

```python
# Qualquer valor indicativo de ausência será ordenado no final da Series, por padrão
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])
```

```python
obj.sort_values()
```

```python
# Ordenando por colunas
frame = pd.DataFrame({'b': [4, 7, -3, 2], 'a': [0, 1, 0, 1]})
```

```python
frame
```

```python
frame.sort_values(by='b')
```

```python
# Ordenando por várias colunas
frame.sort_values(by=['a', 'b'])
```

### Valores únicos, contadores de valores e pertinência

```python
obj = pd.Series(['c', 'a', 'd', 'a', 'a', 'b', 'b', 'c', 'c'])
```

```python
# Valores únicos
uniques = obj.unique()
```

```python
uniques
```

```python
# Ordenando os valores únicos
uniques.sort()
uniques
```

```python
# Contagem de valores
obj.value_counts()
```

```python
obj
```

```python
# Pertinência a conjuntos
mask = obj.isin(['b', 'c'])
```

```python
mask
```

```python
obj[mask]
```

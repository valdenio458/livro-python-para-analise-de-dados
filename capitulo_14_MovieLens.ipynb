## 14.2 Conjunto de dados do MovieLens 1M

Este conjunto de dados contém um milhão de avaliações coletadas de 6 mil usuários sobre 4 mil filmes.
Os dados estão espalhados em três tabelas:
avaliações, informações de usuários e informações sobre filmes.

```python
import pandas as pd
```

```python
# Reduz a exibição
pd.options.display.max_rows = 10
```

### Baixando users.dat

```python
unames = ['user_id', 'gender','age', 'occupation', 'zip']
users = pd.read_table('/content/users.dat', sep='::', header=None, names=unames)
```

```python
import wget
```

```python
url = 'https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/movielens/ratings.dat'
```

```python
# Diretório onde você quer salvar o arquivo
save_path = '/content/'
```

```python
# Baixar o arquivo
wget.download(url, save_path)
```

```python
import os
file_path = os.path.join(save_path, 'ratings.dat')
if os.path.exists(file_path):
    print("Arquivo baixado com sucesso!")
else:
    print("Falha ao baixar o arquivo.")
```

```python
# Leitura do arquivo usando pandas
data = pd.read_csv(file_path, sep='::', engine='python', names=['UserID', 'MovieID', 'Rating', 'Timestamp'])
```

```python
# Visualizar as primeiras linhas do DataFrame
print(data.head())
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(file_path, sep='::',header=None, names =rnames)
```

### Baixando movies.dat

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('/content/movies.dat', sep='::',
                       header=None, names=mnames)
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
# Obter as avaliações médias de cada filme agrupadas pela identificação de gênero
mean_ratings = data.pivot_table('rating',index='title',
                                columns='gender', aggfunc='mean')
```

```python
ratings_by_title = data.groupby('title').size()
```
